# Published Wind Data Plots (ALMA Ground-Based)

This notebook generates visualizations for the ground-based meteorological data downloaded for the ALMA site, including wind roses, time series for speed/gust, and component analysis.

```python
import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from windrose import WindroseAxes

# allow config import from project root
PROJECT_ROOT = os.path.abspath(r"..\..")
sys.path.append(PROJECT_ROOT)

from config import PUBLISHED_DIR, PUBLISHED_COLS

print("CONFIG LOADED ✔")
```

```python
# Load the processed timeseries data
data_path = r'../../data/exports/alma_published_timeseries.csv'
df = pd.read_csv(data_path, parse_dates=['datetime'])
df = df.sort_values('datetime')

print(f"Loaded {len(df)} records.")
df.head()
```

## Wind Speed and Gust Time Series

```python
plt.figure(figsize=(15, 6))
plt.plot(df['datetime'], df['speed'], alpha=0.7, label='Average Speed')
if 'gust' in df.columns:
    plt.plot(df['datetime'], df['gust'], alpha=0.3, label='Wind Gust', color='red')

plt.title('ALMA Ground-Based Wind Speed & Gust (2001-2002)')
plt.xlabel('Date')
plt.ylabel('Wind Speed (m/s)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Wind Rose Analysis

```python
# Drop rows with missing direction for the wind rose
df_rose = df.dropna(subset=['direction', 'speed'])

ax = WindroseAxes.from_ax()
ax.bar(df_rose['direction'], df_rose['speed'], normed=True, opening=0.8, edgecolor='white')
ax.set_legend()
plt.title("ALMA Wind Rose (Published Data)")
plt.show()
```

## U and V Component Distributions

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].hist(df['u'].dropna(), bins=50, color='skyblue', edgecolor='black')
axes[0].set_title('U Component (East-West)')
axes[0].set_xlabel('m/s')

axes[1].hist(df['v'].dropna(), bins=50, color='salmon', edgecolor='black')
axes[1].set_title('V Component (North-South)')
axes[1].set_xlabel('m/s')

plt.tight_layout()
plt.show()
```
