```python
import xarray as xr
import numpy as np
import pandas as pd
import glob
import os

files = sorted(glob.glob(r"C:\Users\2021n\Downloads\era5_wind_data_analysis\data\raw\chile\chile_*.nc"))

print("Files found:", len(files))

results = []

for f in files:
    print(f"\nProcessing {f}")
    
    ds = xr.open_dataset(f)
    
    # --------------------------
    # 1. Compute wind speed
    # --------------------------
    wind_speed = np.sqrt(ds.u10**2 + ds.v10**2)
    
    # --------------------------
    # 2. Spatial average
    # --------------------------
    wind_mean_spatial = wind_speed.mean(dim=["latitude", "longitude"])
    
    # --------------------------
    # 3. Statistics
    # --------------------------
    mean = wind_mean_spatial.mean().item()
    std = wind_mean_spatial.std().item()
    maxv = wind_mean_spatial.max().item()
    
    p50 = wind_mean_spatial.quantile(0.50).item()
    p90 = wind_mean_spatial.quantile(0.90).item()
    p95 = wind_mean_spatial.quantile(0.95).item()
    p99 = wind_mean_spatial.quantile(0.99).item()
    
    # --------------------------
    # 4. Extract date from filename
    # --------------------------
    name = os.path.basename(f)
    period = name.replace("chile_", "").replace(".nc", "")
    
    # --------------------------
    # 5. Store results
    # --------------------------
    results.append({
        "month": period,
        "mean_wind": mean,
        "std_wind": std,
        "max_wind": maxv,
        "p50": p50,
        "p90": p90,
        "p95": p95,
        "p99": p99
    })
    
    ds.close()
```

```python
df = pd.DataFrame(results)

# Sort chronologically
df = df.sort_values("month")

print(df)
```
