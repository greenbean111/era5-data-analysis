```python
import xarray as xr
import pandas as pd

# load files
files = [
    "data/raw/chile/2001JFMA.nc",
    "data/raw/chile/2001MJJA.nc",
    "data/raw/chile/2001SOND.nc",
    "data/raw/chile/2002JFMA.nc",
    "data/raw/chile/2002MJJA.nc",
    "data/raw/chile/2002SOND.nc"
]

ds = xr.open_mfdataset(files, combine="by_coords")

print("Dataset loaded:")
print(ds)

# time
print("\nTime preview:")
print(ds.valid_time.values[:10])
print(ds.valid_time.values[-10:])

# sort by valid time
ds['valid_time'] = pd.to_datetime(ds.valid_time.values)
ds = ds.sortby("valid_time")

# group by year and month
monthly_groups = ds.groupby(ds.valid_time.dt.strftime('%Y-%m'))

print("\nDetected months:")
print(list(monthly_groups.groups.keys()))

# save the months
for period, data in monthly_groups:
    filename = f"chile_{period}.nc"
    
    print(f"\nSaving {filename}")
    print(f"Time range: {data.valid_time.values[0]} -> {data.valid_time.values[-1]}")
    print(f"Timesteps: {data.sizes['valid_time']}")
    
    data.to_netcdf(filename)

#
print("\nVALIDATION CHECK:")
for period, data in monthly_groups:
    months = set(pd.to_datetime(data.valid_time.values).month)
    print(f"{period}: {months}")
```

```python
import glob

files = glob.glob("chile_*.nc")

for f in files:
    ds = xr.open_dataset(f)
    
    print(f"\nChecking {f}")
    print("Timesteps:", ds.sizes["valid_time"])
    
    # Check for existence
    var = list(ds.data_vars)[0]
    
    nan_count = ds[var].isnull().sum().item()
    total = ds[var].size
    
    print(f"{var}: {nan_count}/{total} NaNs")
```
